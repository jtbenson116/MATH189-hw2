## Cleaning

```python
#NaNs
video.isnull().sum()
```

looks like there's one Nan per column, lets find out where

```python
for col in video:
    print(video[col][video[col].isnull()].index)
```

They're all at index 91 so we can drop that column

```python
video = video.drop(91, axis=0)
video.head()
```

now we can address the issue with the first column

```python
split_col = video['time like'].str.replace('\s+',' ').str.split(' ').to_list()
video[['time','like']] = split_col
video['time'] = video.time.astype(float)
video['like'] = video.like.astype(float)
video = video.drop('time like',axis=1)
```
